# **K MEANS CLUSTERING**

This is an algorithm which allows us to classify **unlabelled** data in **unsupervised** learning algorithms. The overall goal is to divide data into distinct groups such that the observations within each group are **similar**. We are not trying to produc any sort of outcome, but we are trying to find patterns in the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
#make_blobs generates blob like data that you can then use to cluster by adding
#cluster standard deviation
```

```python
data=make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data
#this is essentially a tuple
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
```

```python
from sklearn.cluster import KMeans
```

This algorithm (KMeans) <br>
1. randomly assigns each observation in your blob to a cluster and finds the centroid of the cluster.<br>
2. It reassigns data points to the cluster whose centroid is the closest and keeps iterating these two steps. It gets repeated over and over again until the cluster variation cannot be reduced any further.

```python
 kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
data[1]
```

```python
fig, (axes1, axes2)=plt.subplots(1,2, sharey=True, figsize=(10,6))
axes1.set_title("K Means")
axes2.set_title("Original")

axes1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_)
axes2.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
 kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(data[0])
```

```python
fig, (axes1, axes2)=plt.subplots(1,2, sharey=True, figsize=(10,6))
axes1.set_title("K Means")
axes2.set_title("Original")

axes1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_)
axes2.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
kmeans=KMeans(n_clusters=9)
kmeans.fit(data[0])
fig, (axes1, axes2)=plt.subplots(1,2, sharey=True, figsize=(10,6))
axes1.set_title("K Means")
axes2.set_title("Original")

axes1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_)
axes2.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
kmeans=KMeans(n_clusters=15)
kmeans.fit(data[0])
fig, (axes1, axes2)=plt.subplots(1,2, sharey=True, figsize=(10,6))
axes1.set_title("K Means")
axes2.set_title("Original")

axes1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap="rainbow")
axes2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
```

**More the clusters, less the error.**
